# 표준 출력
- 이스케이프 문자
    - \' : 작은따옴표(single quote), '
    - \" : 큰따옴표(double quote), "
    - \n : 줄 바꿈(line feed), LF
    - \r : 캐리지 리턴(carriage return), CR
    - \t : 탭(tab), TAB
    - \v : 수직 탭(vertical tab), VT
    - \\ : 역슬래시
    - \a : 벨(bell), BEL
    - \b : 백스페이스(back space), BS

- print() 함수
    - print(value, ... sep=' ', end='\n', file-sys.stdout, flush=False)
    - sep : 출력할 value의 구분자
        - 기본값 : ' '(의미 : 공백)
        - 사용하지 않고 print()함수 사용하면 출력 대상은 공백으로 구분
    - end : value 출력 후 출력할 문자
        - 기본값 : '\n'(의미 : 줄 바꿈)
        - 사용하지 않고 print()함수 사용하면 출력 대상은 공백으로 구분 
    - file : 출력 방향 지정
        - 기본값 : sys.stdout(의미 : 모니터를 의미)
        - 사용하지 않고 print()함수 사용하면 출력 대상은 모니터에 출력됨
    - flush : flush 유무 지정
        - 기본값 : False(의미 : 스트림 비우지 않음)
        - 사용하지 않고 print()함수 사용하면 출력 후 출력 스트림은 비워지지 않음

```python
#기본예제1
print('Hello \'World\'')
print("Hello \"World\"")
print('*n**\n***')
print('이름\t연락처')
print('제시카\t02-123-4567')
print('마틴\t010-8765-1234')

print()
#기본예제2
print('재미있는', '파이썬')
print('Python', 'Java', 'C', sep=',')

print()

print('영화 타이타닉')
print('평점', end=':')
print('5점')

fos = open('sample.py', mode='wt')
print('print("Hello World")', file=fos)
fos.close()
```

# 형식을 갖춘 문자열
- % 연산자
    - %d : decimal, 정수(10진수)
    - %o : octal, 정수(8진수)
    - %x : hexadecimal, 정수(16진수)
    - %f : floating point, 실수
    - %s : string, 문자열


```python
# 출력 대상의 필드 폭, 소수점 자릿수 조정

# 5자리로 오른쪽 맞춤 출력
print('%5d' % 1)
print('%5d' % 22)
print('%5d' % 333)
print('%5d' % 4444)
print('%5d' % 55555)

# 5자리로 왼쪽 맞춤 출력
print('%-5d' % 1)
print('%-5d' % 22)
print('%-5d' % 333)
print('%-5d' % 4444)
print('%-5d' % 55555)

# 전체 5자리, 소수 이하 2자리 오른쪽 맞춤
print('%5.2f' % 3.66)

# 전체 5자리, 소수 이하 1자리 오른쪽 맞춤
print('%5.1f' % 3.66)

# 전체 5자리, 소수 이하 2자리 왼쪽 맞춤
print('%-5.2f' % 3.66)

# 전체 5자리, 소수 이하 1자리 왼쪽 맞춤
print('%-5.1f' % 3.66)

# 소수 이하 2자리
print('%.2f' % 3.66)

# 소수 이하 1자리
print('%.1f' % 3.66)
```

```python
# 문자열 형식 기호로 정수와 실수 모두 처리 가능
print('%s' % 10)
print('%s' % 3.14)

#어떤 값을 출력할 때 형태의 변화 없이 그대로 출력하고자 한다면 %s를 사용하는 것도 방법
```

```python
#기본예제
name = 'Kai'
print('내 이름은 %s입니다.' % name)

height = 120.5
print('내 키는 %fcm입니다.' % height)

weight = 23.55
print('내 몸무게는 %.1fkg입니다.' % weight)

year, month, day = 2014, 8, 25
print('내 생일은 %d년 %d월 %d일입니다.' % (year, month, day))
```

```python
# format() 매소드
# format() 메소드의 인수로 변수나 값을 표시하고, 해당 값이 표시될 위치를 중괄호({})로 표시하는 방식
print('My name is {}'.format('James'))
print('My name is {0}'.format('James')) # 인수의 인덱스를 지정한 {0}을 사용해도 같은 결과를 얻을 수 있음
print('My name is {name}'.format(name='James')) # {name}과 같은 방식으로 표시할 변수명 지정 가능

# 3개의 결과가 모두 같음
```

```python
# 기본예제
zipcode = '12345'
print('우편번호 :{}'.format(zipcode))

address = "대전광역시 유성구"

print('주소 : {addr}'.format(addr=address))
tel1, tel2, tel3 = '010', '1234', '5678'
print('연락처 : {0}-{1}-{2}'.format(tel1, tel2, tel3))
```

```python
# f-string
name = 'James'
print(f'My name is {name}')

year = 2026
print(f'내년은 {year+1}년 입니다.')
```
