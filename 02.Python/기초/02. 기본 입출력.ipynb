# 표준 출력
- 이스케이프 문자
\' : 작은따옴표(single quote), '
\" : 큰따옴표(double quote), "
\n : 줄 바꿈(line feed), LF
\r : 캐리지 리턴(carriage return), CR
\t : 탭(tab), TAB
\v : 수직 탭(vertical tab), VT
\\ : 역슬래시
\a : 벨(bell), BEL
\b : 백스페이스(back space), BS

```python
#기본예제
print('Hello \'World\'')
print("Hello \"World\"")
print('*n**\n***')
print('이름\t연락처')
print('제시카\t02-123-4567')
print('마틴\t010-8765-1234')
```
