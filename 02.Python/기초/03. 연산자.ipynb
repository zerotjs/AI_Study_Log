# 연산자
- 산술 연산자(숫자 연산)
    - +(덧셈),-(뺄셈), *(곱셈), **(거듭제곱), /(나눗셈), //(몫), %(나머지)
- 대입 연산자(대입 및 복합 대입)
    - =(변수 저장), +=(변수에 우변을 덧셈한 값 저장), -=(변수에 우변을 뺄셈한 값 저장), *=(변수에 우변을 곱한 값 저장), **=(변수에 우변을 거듭제곱한 값 저장), /=(변수에 우변을 나눈 값 저장), //=(변수에 우변을 나눈 몫 저장), %=(변수에 우변을 나눈 나머지 저장)
- 관계 연산자(크기 비교)
    - 2개의 항을 비교하여 그 결과를 논리 자료형으로 반환
    - >, >=, <, <=, ==, !=
- 논리 연산자
    - and(모두 참이여야 True, 아니면 False), or(하나라도 참이면 True, 아니면 False), not
- 비트 연산자(이진 연산)
    - &, |, ^, ~, <<, >>
- 기타 연산자
    - 조건 연산
        - 참 if 조건식 else 거짓
    - 시퀀스 연산
        - in, +, *
