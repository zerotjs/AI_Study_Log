# Day 1 : 벡터와 선형 결합
- **선형 결합(Linear Combination)**이란?
    - 두 벡터 $v$와 $w$가 있을 때, $cv + dw$ 형태로 만드는 것
    - 이 연산을 통해 벡터 공간 전체를 생성(Span)할 수 있음.
    - 예 : $2v - 3w$는 벡터 $v$와 $w$의 $cv + dw$ 형태의 선형결합

- 더할 때는 같은 차원 (뺄셈도 같은 차원)
    - 예 : $v + w$ = [2,1] + [4,3] = [6,4]


```python
import numpy as np
import matplotlib.pyplot as plt

v = np.array([1,2])
w = np.array([3,1])

result = 2*v + 3*w

print(f"벡터 v : {v}")
print(f"벡터 w : {w}")
print(f"선형 결합 결과 (2v + 3w) : {result}")
```
